```python
#import atomic dataset

import pandas as pd

location = "../sloatomic2020/"


#read in the data
df = pd.read_csv(location + "sloatomic_train.tsv", sep="\t", header=0)
```

```python
#regex
import re

#replace (only the ones that are in the same word)
patternx = r'(?i)\bose.*x\b'
patterny = r'(?i)\bose.*y\b'

#go trough df
for index, row in df.iterrows():
	#replace head_event and tail_event
	try:
		df.at[index, 'head_event'] = re.sub(patternx, '[A]', df.at[index, 'head_event'])
	except:
		print("Error in row: ", index)
	try:
		df.at[index, 'tail_event'] = re.sub(patternx, '[A]', df.at[index, 'tail_event'])
	except:
		print("Error in row: ", index)

	try:
		df.at[index, 'head_event'] = re.sub(patterny, '[B]', df.at[index, 'head_event'])
	except:
		print("Error in row: ", index)
	try:
		df.at[index, 'tail_event'] = re.sub(patterny, '[B]', df.at[index, 'tail_event'])
	except:
		print("Error in row: ", index)
```

```python
#replace (the ones where the previous world starts wit ose
patternx = r'(?i)\bose\w*\s+x\b'
patterny = r'(?i)\bose\w*\s+y\b'

#go trough df
for index, row in df.iterrows():
	#replace head_event and tail_event
	try:
		df.at[index, 'head_event'] = re.sub(patternx, '[A]', df.at[index, 'head_event'])
	except:
		print("Error in row: ", index)
	try:
		df.at[index, 'tail_event'] = re.sub(patternx, '[A]', df.at[index, 'tail_event'])
	except:
		print("Error in row: ", index)

	try:
		df.at[index, 'head_event'] = re.sub(patterny, '[B]', df.at[index, 'head_event'])
	except:
		print("Error in row: ", index)
	try:
		df.at[index, 'tail_event'] = re.sub(patterny, '[B]', df.at[index, 'tail_event'])
	except:
		print("Error in row: ", index)
```

```python
#add head_event and tail_event together
#if head_event has ___ add tail_event to at that position, else add tail_event to the end

for index, row in df.iterrows():
	#replace head_event and tail_event
	try:
		if "___" in df.at[index, 'head_event']:
			df.at[index, 'head_event'] = df.at[index, 'head_event'].replace("___", df.at[index, 'tail_event'])
		else:
			df.at[index, 'head_event'] = df.at[index, 'head_event'] + " " + df.at[index, 'tail_event']
	except:
		print("Error in row: ", index)
```

```python
#save the dataframe
df.to_csv(location + "sloatomic_train_fixed.tsv", sep="\t", index=False)
```
